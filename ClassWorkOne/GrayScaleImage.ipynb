```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import misc
%matplotlib inline
```

```python
pic=plt.imread('./source/picture1.jpg')
```

```python
pic.shape
```

```python
#plt.imshow(pic)
```

```python
#YUV亮度灰度化
weight=np.array([0.3,0.59,0.11])
pic1=np.dot(pic,weight)
```

```python
#输出256分辨率的灰度
h,w=pic1.shape
plt.axis('off')
plt.gcf().set_size_inches(w / 100.0, h / 100.0)
plt.subplots_adjust(top=1, bottom=0, left=0, right=1, hspace=0, wspace=0)
plt.imshow(pic1,cmap='gray')
plt.savefig(r'./Target/256.jpg',pad_inches=0,transparent=False,dpi=100)
plt.savefig(r'./Target/256(0.5).jpg',pad_inches=0,transparent=False,dpi=50)
plt.savefig(r'./Target/256(0.25).jpg',pad_inches=0,transparent=False,dpi=25)
plt.savefig(r'./Target/256(0.125).jpg',pad_inches=0,transparent=False,dpi=12.5)
```

```python
pic1.max()
```

```python
pic1.min()
```

```python
#128灰度级
#pic2t=np.zeros(shape=(pic1.shape[0],pic1.shape[1]))
for i in range(0,pic1.shape[0]):
    pic1[i,:]=(128*pic1[i,:])//(pic1.max()-pic1.min())
```

```python
h,w=pic1.shape
plt.axis('off')
plt.gcf().set_size_inches(w / 100.0, h / 100.0)
plt.subplots_adjust(top=1, bottom=0, left=0, right=1, hspace=0, wspace=0)
plt.imshow(pic1,cmap='gray')
plt.savefig(r'./Target/128.jpg',pad_inches=0,transparent=False,dpi=100)
#plt.show()  
```
